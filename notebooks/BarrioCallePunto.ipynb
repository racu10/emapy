```python
import emapy as epy
```

```python
barris = epy.getDatabase('barris', 'json','../data/raw/barris.geojson', '',True, 0, 1, 'cartodb_id')  
```

```python
locationAreaBoundingBox = (41.3248770036,2.0520401001,41.4829908452,2.2813796997)
allStreets = epy.getDatabaseFromOSM('streetsBarcelona', 'way', False, True,locationAreaBoundingBox, '')
```

```python
allData = epy.getDatabaseFromOSM('escuelas', 'amenity', False, True, locationAreaBoundingBox, 'school')
```

```python
barri = barris[1][33]['geometry']['geometry']['coordinates'][0][0]
street = allStreets[1]
point = allData[1]
```

```python
points = []
map = epy.mapCreation(41.3248770036,2.0520401001)

for point in barri:
    points.append(tuple([point[1], point[0]]))
epy.mapAddLine(map,points, "#0000FF", 2.5, 1)
map

```
